```python
def man_bins(var1):
    if 0.1 < var1 <= 10.817:
        return 'A'
    elif 16.817 < var1 <= 33.433:
        return 'B'
    elif 33.433 < var1 <= 50.05:
        return 'C'
    elif 50.05 < var1 <= 66.667:
        return 'D'
    elif 66.667 < var1 <= 83.283:
        return 'E'
    else:
        return 'F'
```

```python
data_df['bins1_manual'] = data_df[['Water_(g)']].apply(lambda x: man_bins(x['Water_(g)']),axis=1)
```

```python
data_df['bins1_manual'].value_counts()
```

```python
# deciding the ranges
cut_bins = [0, 16.817, 33.433, 50.05, 66.667,83.283, 99.9]
data_df['bins2_manual'] = pd.cut(data_df['Water_(g)'], bins=cut_bins)
```

```python
cut_labels_4 = ['A', 'B', 'C', 'D','E','F']
cut_bins = [0, 16.817, 33.433, 50.05, 66.667,83.283, 99.9]
data_df['bins3_manual'] = pd.cut(data_df['Water_(g)'], bins=cut_bins,labels=cut_labels_4)
```
