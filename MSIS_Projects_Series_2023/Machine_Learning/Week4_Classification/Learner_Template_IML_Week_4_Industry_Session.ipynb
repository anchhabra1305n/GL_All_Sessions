## **<font color='blue'>Classification**

Classification in machine learning is a type of supervised learning task where the goal is to categorize or classify input data into predefined classes or categories based on their features. The process involves training a model on a labeled dataset, where each data point is associated with a specific class label. The model learns from this data and then generalizes its knowledge to classify new, unseen data.

## **<font color='blue'>Linear vs Logistic Regression**

**Linear Regression:**

- Linear regression is a supervised learning algorithm used for predicting continuous numerical values.
- It models the relationship between the dependent variable and one or more independent variables by fitting a linear equation to the data points.
- The output of linear regression is a continuous value, which can be positive or negative.

**Logistic Regression:**

- Logistic regression is a supervised learning algorithm used for binary classification tasks.
- It models the probability of a binary outcome (e.g., yes/no, 1/0) by fitting a logistic (sigmoid) function to the data points.
- The output of logistic regression is the probability of belonging to a particular class, which is always between 0 and 1.

**Main Difference:**
-The main difference between linear regression and logistic regression lies in their output and application. Linear regression is used for regression tasks, where the goal is to predict continuous numeric values, whereas logistic regression is used for binary classification tasks, where the goal is to predict the probability of belonging to one of the two classes.


## **<font color='blue'>Logistic Regression**

**What is Logistic Regression?**

- Logistic Regression is a statistical method used for binary classification tasks, where the outcome variable is categorical with two classes (e.g., Yes/No, 1/0, True/False).
- Despite its name, it is a classification algorithm and not a regression algorithm.

**Purpose of Logistic Regression:**

- The main goal of Logistic Regression is to estimate the probability of a sample belonging to a particular class.
- It models the relationship between the independent variables (features) and the probability of the binary outcome.

**Sigmoid Function (Logistic Function):**

## **<font color='blue'>Evaluating Logistic Regression**

Evaluating a logistic regression model involves assessing its performance and effectiveness in making binary classification predictions. There are several metrics and techniques used to evaluate a logistic regression model. Let's explore the main evaluation methods:

- **Confusion Matrix:**
The confusion matrix is a table that summarizes the model's performance by showing the number of true positive (TP), true negative (TN), false positive (FP), and false negative (FN) predictions.

image.png

- **Accuracy:**
Accuracy is the most straightforward metric, representing the percentage of correct predictions over the total number of samples.

  It's calculated as (TP + TN) / (TP + TN + FP + FN).

- **Precision:**
Precision measures the proportion of true positive predictions among all positive predictions. It's calculated as TP / (TP + FP). Precision focuses on the correctness of positive predictions.

- **Recall** (Sensitivity or True Positive Rate):
Recall is the proportion of true positive predictions among all actual positive samples. It's calculated as TP / (TP + FN). Recall focuses on capturing all positive cases.

- **F1 Score:**
The F1 score is the harmonic mean of precision and recall. It's a balanced metric that combines both precision and recall. It's calculated as 2 * (Precision * Recall) / (Precision + Recall).

- **Receiver Operating Characteristic (ROC) Curve:**
The ROC curve is a graphical representation of the model's performance at various classification thresholds. It plots the true positive rate (recall) against the false positive rate (1-specificity) as the threshold varies.

- **Area Under the ROC Curve (AUC-ROC):**
The AUC-ROC is a single metric that quantifies the overall performance of the logistic regression model. It measures the area under the ROC curve. Higher AUC-ROC values indicate better model performance.

- **Precision-Recall Curve:**
The precision-recall curve is another graphical representation that shows the trade-off between precision and recall at different classification thresholds.

The choice of evaluation metrics depends on the specific problem and the importance of correctly identifying positive or negative cases. For instance, in medical diagnosis, recall might be more critical to minimize false negatives (missing positive cases), while in fraud detection, precision might be more crucial to reduce false positives (avoiding false alarms).

It's essential to use a combination of these evaluation methods to gain a comprehensive understanding of the model's performance and make informed decisions about its suitability for the task at hand.

## **<font color='blue'>Discriminant Analysis**

**Discriminant Analysis** (DA) is a classification technique used to distinguish between multiple classes based on the values of predictor variables (features). It seeks to find a decision boundary that best separates the classes in the feature space. There are two main types of Discriminant Analysis: Linear Discriminant Analysis (LDA) and Quadratic Discriminant Analysis (QDA).

**Linear Discriminant Analysis (LDA):**

LDA assumes that the class-conditional probability distributions of the features are multivariate Gaussian and that the covariance matrices of all classes are equal (Homoscedasticity). LDA projects the data onto a lower-dimensional space by finding a linear combination of features that maximizes the separation between the classes while minimizing the scatter within each class.

- Assumes that the covariance matrices are equal for all classes (spherical classes).
- Projects data points onto a linear subspace.
- Works well when classes are well-separated and the data is Gaussian and homoscedastic.

**Quadratic Discriminant Analysis (QDA):**
QDA relaxes the assumption of equal covariance matrices for all classes present in LDA. It allows each class to have its own covariance matrix. As a result, QDA creates a quadratic decision boundary instead of a linear one, making it more flexible in modeling complex data distributions.

- Does not assume equal covariance matrices for all classes (allows for different shapes of classes).
- Projects data points onto a quadratic subspace.
- Well-suited for datasets where classes have different covariances and may exhibit non-linear boundaries.

**Differences between LDA and QDA:**

**Assumptions:**
- LDA assumes equal covariance matrices for all classes (homoscedasticity), while QDA allows each class to have its own covariance matrix (heteroscedasticity).

**Decision Boundary:**
- LDA creates a linear decision boundary, whereas QDA creates a quadratic decision boundary.

**Complexity:**
- LDA is computationally simpler and more efficient since it involves calculating a common covariance matrix. In contrast, QDA requires the estimation of separate covariance matrices for each class, making it computationally more expensive.

**Use Cases:**

- LDA is preferred when classes have similar covariance matrices and the data is relatively simple and can be separated linearly.
- QDA is more suitable for datasets where the classes have different covariances and the data distribution is more complex, allowing for non-linear decision boundaries.


It's essential to choose between LDA and QDA based on the specific characteristics of the dataset and the underlying assumptions about the data distribution. Both methods are powerful tools for classification tasks and are commonly used in various fields, including pattern recognition, machine learning, and statistics.

## **<font color='blue'>Naive Bayes**

Naive Bayes is a probabilistic classification algorithm based on Bayes' theorem, which provides a simple and efficient way to classify data into different classes.


The Naive Bayes algorithm uses Bayes' theorem to calculate the posterior probability of a class given observed features. The formula for Naive Bayes can be expressed as follows:

    P(C|X) = (P(X|C) * P(C)) / P(X)

Where:

- P(C|X) is the posterior probability of class C given the observed features X. This represents the probability of the data point belonging to class C given the features X.
- P(X|C) is the likelihood probability of observing the features X given the class C. It represents the probability of the features X occurring in data points belonging to class C.
- P(C) is the prior probability of class C. This represents the probability of a data point belonging to class C without considering any features.
- P(X) is the probability of observing the features X. It acts as a normalizing factor and ensures that the posterior probabilities sum up to 1 over all classes.

The Naive Bayes algorithm calculates the posterior probabilities for each class and assigns the data point to the class with the highest posterior probability, making it the predicted class.

The "naive" assumption in Naive Bayes is that all features in X are independent of each other given the class C. This simplifies the calculation of the likelihood probability P(X|C), as it can be computed as the product of the individual feature probabilities:

    P(X|C) = P(X_1|C) * P(X_2|C) * ... * P(X_n|C)

Where X_1, X_2, ..., X_n are the individual features in X.

By applying Bayes' theorem and the "naive" assumption, Naive Bayes efficiently estimates the probabilities and effectively classifies data points into different classes based on their observed features.

## **<font color='blue'>K-Nearest Neighbors (KNN)**

K-Nearest Neighbors (KNN) is a simple and intuitive machine learning algorithm used for both classification and regression tasks. It is a non-parametric algorithm, meaning it doesn't make assumptions about the underlying data distribution. Instead, it relies on the similarity between data points to make predictions.

image.png

**K and Distance Metric Selection**:

Choose the value of K, which represents the number of nearest neighbors that will influence the prediction. It's usually an odd number to avoid ties.
Select an appropriate distance metric, such as Euclidean distance or Manhattan distance, to measure the similarity between data points.

**Prediction for a New Data Point:**

Given a new data point (with unknown class), find the K nearest neighbors to this data point based on the chosen distance metric.
Count the number of neighbors belonging to each class.

**Classification (KNN for Classification):**

For classification tasks, assign the new data point to the class that has the most occurrences among its K nearest neighbors. In case of a tie, one can use majority voting or other tie-breaking methods.

**Regression (KNN for Regression):**

For regression tasks, predict the target value for the new data point by taking the average (or weighted average) of the target values of its K nearest neighbors.

## **<font color='blue'>Support Vector Machines**

Support Vector Machine (SVM) is a powerful and versatile supervised machine learning algorithm used for both classification and regression tasks. It is particularly effective in tasks where the data is not linearly separable, as SVM can find complex decision boundaries that maximize the margin between different classes.

**Linear Separation:**

For datasets that are linearly separable (i.e., the classes can be perfectly separated by a straight line or hyperplane), SVM finds the line or hyperplane that maximizes the margin between the two classes.

**Non-Linear Separation:**

For datasets that are not linearly separable, SVM uses a technique called the "kernel trick." It maps the data into a higher-dimensional space where the classes become separable, even if they were not separable in the original space.

**Margin and Support Vectors:**

The margin is the distance between the decision boundary (separating line or hyperplane) and the nearest data points of each class. SVM aims to maximize this margin to ensure better generalization and robustness to new data.
Support vectors are the data points that lie closest to the decision boundary. They are crucial for defining the decision boundary and, therefore, for making predictions.

**C Parameter and Soft Margin:**

SVM introduces a parameter called "C" that controls the trade-off between maximizing the margin and allowing some misclassifications. A large C value results in a hard margin, where SVM tries to classify all training points correctly, even if it leads to a smaller margin. A small C value allows more misclassifications but results in a wider margin.

**Kernel Trick:**

The kernel trick is a mathematical technique that allows SVM to implicitly compute the dot product of data points in the higher-dimensional space without explicitly transforming the data. This makes SVM computationally efficient even in high-dimensional spaces.

- SVM is effective in handling both linearly separable and non-linearly separable data.

- It aims to find the decision boundary that maximizes the margin and generalizes well to new data.

- SVM works well for datasets with small to medium-sized samples.

Remember that SVM's performance depends on the choice of the kernel function, the regularization parameter C, and the preprocessing of the data. Overall, SVM is a versatile and robust algorithm that can achieve high accuracy and generalization in a wide range of machine learning tasks.

### **K-Nearest Neighbors (K-NN)**

K-NN uses features from the training data to predict the values of new data points, which means the new data point will be assigned a value based on how similar it is to the data points in the training set.


The following steps are performed in K-NN:

- Select K
- Calculate distance (Euclidean, Manhattan, etc.)
- Find the K closest neighbors
- Take majority vote for labels

The “K” in the K-NN algorithm is the number of nearest neighbors we wish to take the vote from. Generally, K is taken to be an odd number when the number of classes is even, so as to get a majority vote. Let's say K=3. In that case, we will make a circle with the new data point as the center just as big as enclosing only the three nearest data points on the plane.

**But before actually building the model, we need to identify the value of K to be used in K-NN. We will perform the following steps for the same.**

- For every value of K (from 1 to 15), split the training set into a new train and validation sets (30 times)
- Scale the training data and the validation data
- Take the average of the error on these training and the validation sets for each value of K
- Plot the average train vs validation error for all Ks
- Choose the optimal K from the plot where the two errors are comparable

```python
knn = KNeighborsClassifier()

# We select the optimal value of K for which the error rate is the least in the validation data
# Let us loop over a few values of K to determine the optimal value of K

train_error = []

test_error = []

knn_many_split = {}

error_df_knn = pd.DataFrame()

features = X.columns

for k in range(1, 15):
    train_error = []

    test_error = []

    lista = []

    knn = KNeighborsClassifier(n_neighbors = k)

    for i in range(30):
        x_train_new, x_val, y_train_new, y_val = train_test_split(x_train, y_train, test_size = 0.20)

        # Fitting K-NN on the training data
        knn.fit(x_train_new, y_train_new)

        # Calculating error on the training data and the validation data
        train_error.append(1 - knn.score(x_train_new, y_train_new))

        test_error.append(1 - knn.score(x_val, y_val))

    lista.append(sum(train_error)/len(train_error))

    lista.append(sum(test_error)/len(test_error))

    knn_many_split[k] = lista

knn_many_split
```

```python
kltest = []

vltest = []

for k, v in knn_many_split.items():
    kltest.append(k)

    vltest.append(knn_many_split[k][1])

kltrain = []

vltrain = []

for k, v in knn_many_split.items():
    kltrain.append(k)

    vltrain.append(knn_many_split[k][0])

# Plotting K vs Error
plt.figure(figsize = (10, 6))

plt.plot(kltest, vltest, label = 'test' )

plt.plot(kltrain, vltrain, label = 'train')

plt.legend()

plt.show()
```

**Observations:**
- We can see that the test error (error on the validation data) is more or less similar for K greater than or equal to 5. But the training error keeps increasing with increasing K.
- This implies that we would get a lower train and test error if we choose K = 5. Also, if we choose a higher value of K, the model would get biased due to the imbalance in the dataset.
- So, let's fit the K-NN model with **K=5** on the entire training set.

```python
# Define K-NN model

knn = KNeighborsClassifier(n_neighbors = 5)
```

### **Using GridSearchCV for Hyperparameter tuning of the model**

* Hyperparameter tuning is tricky in the sense that there is no direct way to calculate how a change in the hyperparameter value will reduce the loss of your model, so we usually resort to experimentation.
* **Grid search** is a model tuning technique that attempts to compute the optimum values of hyperparameters.
* It is an exhaustive search that is performed on specific parameter values of a model.
* The parameters of the estimator/model used to apply these methods are optimized by cross-validated grid-search over a parameter grid.

- **n_neighbors**

    - Number of neighbors to use.


- **weights={'uniform', 'distance'}**
    - uniform : uniform weights. All points in each neighborhood are weighted equally.
    - distance : weight points by the inverse of their distance. In this case, the closest neighbors of a query point will have a greater influence than neighbors that are further away.


- **p**
    - When p = 1, this is equivalent to using Manhattan_distance (L1), and Euclidean_distance (L2) is used for p = 2.

```python
params_knn = {'n_neighbors': np.arange(3, 15), 'weights': ['uniform', 'distance'], 'p': [1, 2]}

grid_knn = GridSearchCV(estimator = knn, param_grid = params_knn, scoring = 'recall', cv = 10)

model_knn = grid_knn.fit(x_train,y_train)

knn_estimator = model_knn.best_estimator_

print(knn_estimator)
```

```python
# Fit the best estimator on the training data
knn_estimator.fit(x_train, y_train)
```

```python
y_pred_test_knn_estimator = knn_estimator.predict(x_test)

metrics_score(y_test, y_pred_test_knn_estimator)
```

### **Linear Discriminant Analysis**

Linear discriminant analysis (LDA) is generally used to classify patterns between two classes; however, it can be extended to classify multiple patterns. LDA assumes that all classes are linearly separable and according to this, multiple linear discrimination functions representing several hyperplanes in the feature space are created to distinguish between the classes. If there are two classes, then the LDA draws one hyperplane and projects the data onto this hyperplane in such a way as to maximize the separation of the two categories. This hyperplane is created according to two criteria considered simultaneously:

- Maximizing the distance between the means of two classes.
- Minimizing the variation between each category.
