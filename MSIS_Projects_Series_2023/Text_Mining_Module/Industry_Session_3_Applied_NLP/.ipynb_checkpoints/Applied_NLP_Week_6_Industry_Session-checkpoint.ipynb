<font color=darkblue><h2>**Let's recollect some core concepts of Paradigm Shift in NLP**</h2></font>

A paradigm shift refers to a fundamental change in the way something is perceived, understood, or approached. In the field of Natural Language Processing (NLP), there have been several paradigm shifts that have significantly impacted the way NLP tasks are tackled. Here are a few notable paradigm shifts in NLP:

**1.Rule-based to Statistical Approaches:** NLP shifted from manually crafting rules to using statistical models and machine learning algorithms to automatically learn patterns and extract information from data.

**2.Supervised Learning to Unsupervised and Semi-Supervised Learning:** NLP moved from relying heavily on labeled data to leveraging unsupervised and semi-supervised learning methods, allowing models to learn from unlabeled data and generalize to different tasks.

**3.Linguistic Rules to Neural Networks:** Traditional rule-based approaches were replaced by neural networks, enabling end-to-end learning without the need for explicit rule-based feature engineering.

**4.Symbolic to Distributional Semantics:** NLP transitioned from relying on symbolic representations to using distributional semantics, capturing word meanings based on their distributional properties in large corpora.

**5.Task-Specific to General-Purpose Models:** Instead of creating separate models for each task, there has been a shift towards developing general-purpose models that can be fine-tuned for various NLP tasks, leading to more versatile and efficient systems.


These paradigm shifts have significantly influenced the advancements in NLP, allowing for more powerful and flexible language processing systems with improved performance on various tasks.

<font color=darkblue><h2>**Sequence-to-sequence models**</h2></font>

image.png   

Image Source: [Seq-t0-Seq model architecture](https://www.linkedin.com/pulse/anatomy-sequence-to-sequence-machine-translation-rnn-gru-sobh-phd/)

**The trends of paradigm shifting in NLP involve using sequence-to-sequence (Seq-to-Seq) models for different tasks. This means treating tasks like text classification, text matching, reading comprehension, and sequence labeling as problems where the input and output are both sequences of words. Seq-to-Seq models capture context, dependencies, and relationships in the data, resulting in improved performance compared to traditional approaches. This paradigm shift allows for a more comprehensive and context-aware understanding of text data.**

**1.Text Classification as Seq-to-Seq:** Instead of assigning a single label to a document, Seq-to-Seq models predict labels for each word in the document, capturing contextual information.

**2.Text Matching as Seq-to-Seq:** Seq-to-Seq models compare pairs of texts by treating them as sequences, learning complex relationships and semantic similarities.

**3.Reading Comprehension as Seq-to-Seq:** Seq-to-Seq models answer questions based on passages by generating sequences as answers, considering contextual dependencies.

**4.Sequence Labeling as Seq-to-Seq:** Instead of labeling words individually, Seq-to-Seq models predict labels for each word in a sequence, leveraging contextual information for improved accuracy.

These trends shift the paradigm towards using Seq-to-Seq models, allowing for better context understanding, capturing dependencies, and improving performance in these NLP tasks.

<font><h3>**Sample codes for Pre-trained BERT model for Sequence Classification**</h3></font>

image.png

Image Source: [BERT model](https://www.exxactcorp.com/blog/Deep-Learning/how-do-bert-transformers-work)

**BERT (Bidirectional Encoder Representations from Transformers) is a transformer-based model used for natural language processing (NLP) tasks. It captures the contextual meaning of words by considering the surrounding words from both directions. BERT is pre-trained on large amounts of text data using unsupervised learning and can be fine-tuned for specific NLP tasks. It has achieved state-of-the-art performance in various tasks such as text classification, named entity recognition, and question answering. BERT's effectiveness and versatility have made it a widely adopted model in the NLP community.**

In this code, we fine-tune a pre-trained BERT model for sequence classification. We first load the pre-trained tokenizer and model from the Hugging Face's Transformers library. Next, we prepare our data, tokenize the input texts using the tokenizer, and create a PyTorch dataset and data loader.

We set up the necessary components for training, such as the optimizer (AdamW) and the loss function (CrossEntropyLoss). We iterate over the data loader and perform forward and backward passes to train the model. Finally, we print the average loss for each epoch.

This paradigm shift in NLP, using transformer-based models, has revolutionized various NLP tasks, including machine translation, question answering, sentiment analysis, and more, achieving state-of-the-art performance on multiple benchmarks.

```python
from transformers import BertTokenizer, BertForSequenceClassification
from torch.utils.data import DataLoader
from torch.nn import CrossEntropyLoss
import torch

# Load pre-trained tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForSequenceClassification.from_pretrained('bert-base-uncased')

# Prepare data
texts = ["This is a positive sentence.", "This is a negative sentence."]
labels = [1, 0]

# Tokenize input texts
inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

# Create dataset
dataset = torch.utils.data.TensorDataset(inputs['input_ids'], inputs['attention_mask'], torch.tensor(labels))

# Create data loader
batch_size = 2
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Set optimizer and loss function
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)
loss_fn = CrossEntropyLoss()

# Training loop
epochs = 5
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=1, gamma=0.1)

for epoch in range(epochs):
    total_loss = 0
    model.train()
    for step, batch in enumerate(dataloader):
        input_ids, attention_mask, batch_labels = batch
        input_ids, attention_mask, batch_labels = input_ids.to(device), attention_mask.to(device), batch_labels.to(device)

        # Zero gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=batch_labels)
        loss = outputs.loss

        # Backward pass
        loss.backward()

        # Update weights
        optimizer.step()

        total_loss += loss.item()

    scheduler.step()
    print(f"Epoch {epoch + 1}: Average Loss = {total_loss / len(dataloader):.4f}")
```

##### Note: If you want to use the pre-trained model for making predictions or inferences on your specific task, you need to train or "fine-tune" the model on your task-specific dataset. The pre-trained model serves as a starting point, but it needs further training on your data to be effective for your specific task. Since we are training our model on only two sentences loss is high.

<font><h3>**Sample codes for Pre-trained GPT-2 model to generate text based on an input prompt**</h3></font>

Image Source: [GPT-2](https://www.researchgate.net/figure/Architecture-of-the-GPT-2-Transformer-model_fig2_345654307)

**GPT-2 is a powerful language model developed by OpenAI. It is trained on a large amount of text data and can generate high-quality text based on a given prompt. It is known for its impressive ability to generate coherent and contextually relevant text. GPT-2 has been used in various applications such as text completion, generation, and translation. It is widely regarded for its creative and accurate text generation capabilities.**

The provided code demonstrates how to use the Hugging Face Transformers library to load the GPT-2 tokenizer and model, encode an input text, and generate text based on the input using the GPT-2 model.

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load the GPT-2 tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Encode input text
input_text = "Hi, how are you"
input_ids = tokenizer.encode(input_text, return_tensors="pt")

# Generate text based on input
attention_mask = torch.ones_like(input_ids)
output = model.generate(input_ids, attention_mask=attention_mask, max_length=100, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)

# Decode and print the generated text
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```

<font><h3>**Some more examples for different text classification and text matching tasks using sequence-to-sequence (Seq-to-Seq) models:**</h3></font>



**Text Classification as Seq-to-Seq:**

```python
# Load the necessary libraries and models
from transformers import TFGPT2ForSequenceClassification, GPT2Tokenizer
import tensorflow as tf

# Load the GPT-2 tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

# Prepare the input text
input_text = "Text Classification Example."

# Tokenize and encode the input text
input_ids = tokenizer.encode(input_text, return_tensors='tf')

# Load the GPT-2 for sequence classification model
model = TFGPT2ForSequenceClassification.from_pretrained('gpt2')

# Perform the sequence classification
outputs = model(input_ids)

# Get the predicted class probabilities
probs = tf.nn.softmax(outputs.logits, axis=-1)

# Print the predicted class probabilities
print(probs)
```

**Text Matching as Seq-to-Seq:**

```python
from transformers import TFGPT2Model, GPT2Tokenizer
import tensorflow as tf

# Load the GPT-2 tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

# Add a padding token
tokenizer.add_special_tokens({'pad_token': '[PAD]'})

# Example input sentences
sentence1 = "This is the first sentence."
sentence2 = "This is the second sentence."

# Tokenize and encode the input sentences
inputs = tokenizer([sentence1, sentence2], padding=True, truncation=True, return_tensors='tf')

# Retrieve the input IDs and attention mask
input_ids = inputs['input_ids']
attention_mask = inputs['attention_mask']

# Create the GPT-2 model
model = TFGPT2Model.from_pretrained('gpt2')

# Perform forward pass
outputs = model(input_ids, attention_mask=attention_mask)

# Get the last hidden state of the model
last_hidden_state = outputs.last_hidden_state

# Print the last hidden state
print(last_hidden_state)
```

**Reading Comprehension as Seq-to-Seq:**

```python
from transformers import BertTokenizer, TFBertForQuestionAnswering
import tensorflow as tf

# Load the pre-trained model and tokenizer
model_name = "bert-base-uncased"
model = TFBertForQuestionAnswering.from_pretrained(model_name)
tokenizer = BertTokenizer.from_pretrained(model_name)

# Fine-tuning on a question-answering dataset (replace with your own fine-tuning data)
# ...

# Example input
context = "The quick brown fox jumps over the lazy dog."
question = "What does the fox do?"

# Tokenize the input
inputs = tokenizer.encode_plus(question, context, truncation=True, padding='max_length', max_length=512, return_tensors="tf")

# Get the predicted answer
outputs = model(inputs)
start_logits = outputs.start_logits
end_logits = outputs.end_logits

# Decode the predicted answer
all_tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])
start_index = tf.argmax(start_logits, axis=1).numpy()[0]
end_index = tf.argmax(end_logits, axis=1).numpy()[0]
answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][start_index:end_index+1]))

print("Question:", question)
print("Answer:", answer)
```

**Sequence Labeling as Seq-to-Seq:**

```python
#from transformers import BertForTokenClassification, BertTokenizer
#from tensorflow.python.ops.numpy_ops import np_config
#np_config.enable_numpy_behavior()

import tensorflow as tf
from transformers import TFBertForTokenClassification, BertTokenizer

# Load the BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Prepare the input text
input_text = "This is an example sentence."

# Tokenize and encode the input text
input_ids = tokenizer.encode(input_text, add_special_tokens=True, truncation=True, padding='max_length', max_length=128, return_tensors='tf')

# Load the pre-trained BERT for token classification model
model = TFBertForTokenClassification.from_pretrained('bert-base-uncased')

# Perform the token classification
outputs = model(input_ids)

# Get the predicted labels
predicted_labels = tf.argmax(outputs.logits, axis=-1)

# Decode and print the predicted labels
decoded_labels = tokenizer.decode(predicted_labels[0])
print(decoded_labels)
```

**The transformer architecture has several advantages over previous NLP models:**

**Parallelization:** Transformers can process all input tokens simultaneously, making them highly parallelizable and efficient, which is particularly beneficial for tasks involving long sequences.

**Contextual Understanding:** The self-attention mechanism allows transformers to capture the context and dependencies between words effectively, leading to improved understanding of the input sequence.

**Transfer Learning:** Pretrained transformer models, such as BERT (Bidirectional Encoder Representations from Transformers) and GPT (Generative Pre-trained Transformer), have become popular. They are trained on large-scale language modeling tasks and can be fine-tuned for a wide range of downstream NLP tasks, enabling transfer learning and achieving state-of-the-art results with less labeled data.

**Interpretability:** Transformers provide interpretability through attention weights, indicating which words or tokens are most important for making predictions. This can help in understanding the model's decision-making process.

<font><h2>**BERT Model**</h2></font>

```python
#  Sample data of "20 Newsgroups" dataset

from sklearn.datasets import fetch_20newsgroups

# Fetch the dataset
data = fetch_20newsgroups()

# Print the target names
print(data.target_names)

# Print the first document
print(data.data[0])

# Print the target label of the first document
print(data.target[0])
```

```python
import transformers
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from transformers import BertTokenizer, TFBertForSequenceClassification, GPT2Tokenizer, TFGPT2LMHeadModel
import tensorflow as tf

# Load the 20 Newsgroups dataset
newsgroups_dataset = fetch_20newsgroups(subset='all')

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(newsgroups_dataset.data, newsgroups_dataset.target, test_size=0.2, random_state=42)

# BERT Model

# Load the BERT tokenizer
tokenizer_bert = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenize and encode the training data
X_train_encoded = tokenizer_bert.batch_encode_plus(X_train, padding=True, truncation=True, return_tensors='tf')
X_train_input_ids = X_train_encoded['input_ids']
X_train_attention_mask = X_train_encoded['attention_mask']

# Tokenize and encode the testing data
X_test_encoded = tokenizer_bert.batch_encode_plus(X_test, padding=True, truncation=True, return_tensors='tf')
X_test_input_ids = X_test_encoded['input_ids']
X_test_attention_mask = X_test_encoded['attention_mask']

# Create the BERT model
model_bert = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=20)

# Compile the BERT model
optimizer_bert = tf.keras.optimizers.Adam(learning_rate=2e-5)
loss_bert = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
metric_bert = tf.keras.metrics.SparseCategoricalAccuracy('accuracy')
model_bert.compile(optimizer=optimizer_bert, loss=loss_bert, metrics=[metric_bert])

# Reduce Batch Size, system might crash due to high batch size
batch_size = 8

# Train the BERT model
history_bert = model_bert.fit(
    [X_train_input_ids, X_train_attention_mask],
    y_train,
    validation_data=([X_test_input_ids, X_test_attention_mask], y_test),
    epochs=3,
    batch_size=batch_size
)
```

- The model's accuracy improved over the epochs, indicating that it learned to make better predictions.
