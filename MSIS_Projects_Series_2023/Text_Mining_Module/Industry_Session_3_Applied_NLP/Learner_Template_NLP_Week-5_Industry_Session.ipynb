### **<font color='blue'> n-gram Language Model**

Language models determine the probability distribution of word sequences. It is often assumed that a word's probability depends only on the N words that come before it. This is called n-gram Language Model. The counts of all observed word sequences of lengths ranging from one (referred to as a "unigram") to lengths N are saved in an N-Gram model of order N. If an N-gram sequence is queried during inference that was not observed during training, the sequence is then reduced to the probability of the N-1 last words, weighted by an estimated backoff probability.

If the number of words is

- 1 - Unigram
- 2 - Bigram
- 3 - Trigram
- n - ngram

Thr probability of unigram is just a probability of the word in a document or corpus and is given by

**P**(***word***/***count of words in corpus***)

For words more than 1 (i.e Bigram to ngram) work based on the conditional probability of a word given the occurrence of n-1 previous words in a sequence.


### **<font color='blue'> Autoregressive Language Model**

Autoregressive language models (AR) is one of the language modeling methods. Based on the grouping of words that came before it, it predicts the next word in the sentence.

The AR language model has seen significant development over time and has proven to be more effective than conventional language processing techniques.

In its most basic form, the autoregressive language model works by reading through enormous amounts of text and predicting specific words that fit the word combination previously entered. This also called predictive analysis, and it will help you comprehend the language better. It used to take hours to write code with this degree of precision and comprehension. The activities are now easier, more effective, and quicker because to the development of autoregressive language modelling.

### **<font color='blue'> Transfer Learning**

With the help of transfer learning, you can apply or transfer the knowledge acquired from one set of tasks and/or domains to another set of tasks and/or domains. This means that a model that was previously open sourced and trained with a lot of resources including data, processing power, time, and money can now be improved upon and used in new contexts by the larger engineering community with a fraction of the initial resource needs.

Transfer learning refers to the process of taking information from one setting— `the source` —and applying it to another— `the target`.

### **<font color='blue'> GPT vs BERT**

The first pretrained Transformer model, **GPT**, was employed for fine-tuning on numerous NLP tasks and produced cutting-edge outcomes.

Another large pretrained model, **BERT**, was created with the aim of creating better sentence summaries.

The architecture of GPT and BERT clearly distinguishes them from one another. GPT is an autoregressive model while BERT is bidirectional. BERT considers both the left and right contexts when formulating predictions, whereas GPT solely takes the left context into account. As a result, BERT is better suited for tasks like sentiment analysis or natural language understanding (NLU), where knowing the whole context of a sentence or phrase is crucial.

##**<font color='blue'> How can I leverage State-of-the-Art Natural Language Models with only one line of code ?**

Introduced in transformers v2.3.0, **pipelines** provides a high-level, easy to use,
API for doing inference over a variety of downstream-tasks, including:

- ***Sentence Classification _(Sentiment Analysis)_***: Indicate if the overall sentence is either positive or negative, i.e. *binary classification task* or *logitic regression task*.
- ***Token Classification (Named Entity Recognition, Part-of-Speech tagging)***: For each sub-entities _(*tokens*)_ in the input, assign them a label, i.e. classification task.
- ***Question-Answering***: Provided a tuple (`question`, `context`) the model should find the span of text in `content` answering the `question`.
- ***Mask-Filling***: Suggests possible word(s) to fill the masked input with respect to the provided `context`.
- ***Summarization***: Summarizes the ``input`` article to a shorter article.
- ***Feature Extraction***: Maps the input to a higher, multi-dimensional space learned from the data.

Pipelines encapsulate the overall process of every NLP process:

 1. *Tokenization*: Split the initial input into multiple sub-entities with ... properties (i.e. tokens).
 2. *Inference*: Maps every tokens into a more meaningful representation.
 3. *Decoding*: Use the above representation to generate and/or extract the final output for the underlying task.

The overall API is exposed to the end-user through the `pipeline()` method with the following
structure:

```python
from transformers import pipeline

# Using default model and tokenizer for the task
pipeline("<task-name>")

# Using a user-specified model
pipeline("<task-name>", model="<model_name>")

# Using custom model/tokenizer as str
pipeline('<task-name>', model='<model name>', tokenizer='<tokenizer_name>')
```

## <font color='blue'> Data Description


The dataset used for this case study is taken from **Kaggle **website

###**Data Set**

The labeled data set consists of 50,000 IMDB movie reviews, specially selected for sentiment analysis. The sentiment of reviews is binary, meaning the IMDB rating < 5 results in a sentiment score of 0, and rating >=7 have a sentiment score of 1. No individual movie has more than 30 reviews. The 25,000 review labeled training set does not include any of the same movies as the 25,000 review test set. In addition, there are another 50,000 IMDB reviews provided without any rating labels.

###**File descriptions**

- **labeledTrainData** - The labeled training set. The file is tab-delimited and has a header row followed by 25,000 rows containing an id, sentiment, and text for each review.  
- **testData** - The test set. The tab-delimited file has a header row followed by 25,000 rows containing an id and text for each review. Your task is to predict the sentiment for each one.
- **unlabeledTrainData** - An extra training set with no labels. The tab-delimited file has a header row followed by 50,000 rows containing an id and text for each review.
- **sampleSubmission** - A comma-delimited sample submission file in the correct format.

###**Data fields**
- **id** - Unique ID of each review
- **sentiment** - Sentiment of the review; 1 for positive reviews and 0 for negative reviews
- **review** - Text of the review

## <font color='blue'> Data Processing

**Load data**

````python
import tensorflow as tf
import pandas as pd
````

````python
df = pd.read_csv('./labeledTrainData.tsv.zip', sep='\t')
print(df.shape)
````

````python
df.sample(n=5)
````

````python
#Sentences and labels
sentences = df.review.values
labels = df.sentiment.values
````

**Tokenize data using Bert Tokenizer**

````python
!pip install sentencepiece
````

````python
from transformers import *
````

````python
#Get BertTokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
````

````python
#tokenizer.vocab.items()
````

````python
tokenized_texts = [tokenizer.tokenize(sent) for sent in sentences]
````

````python
sentences[0]
````

````python
#Check tokenized text
print(tokenized_texts[0])
````

````python
#We will use only first 200 tokens to do classification (this value can be changed)
max_length = 200
tokenized_texts = [sent[:max_length] for sent in tokenized_texts]
````

````python
for i in range(len(tokenized_texts)):
    sent = tokenized_texts[i]
    sent = ['[CLS]'] + sent + ['[SEP]']
    tokenized_texts[i] = sent
````

````python
print(tokenized_texts[0])
````

````python
#Convert tokens into IDs
input_ids = [tokenizer.convert_tokens_to_ids(sent) for sent in tokenized_texts]
````

````python
print(input_ids[0])
````

````python
tokenizer.vocab.items()
````

````python
#Pad our tokens which might be less than max_length size
input_ids = tf.keras.preprocessing.sequence.pad_sequences(input_ids, maxlen=max_length+2, truncating='post', padding='post')
````

````python
input_ids.shape
````

**Split data between training and test**

````python
from sklearn.model_selection import train_test_split
````

````python
#80% data will be used for training while 20% will be used for test
trainX, testX, trainY, testY = train_test_split(input_ids, labels, test_size=0.2, random_state=12345)
````

**Create Attention masks :** Attention masks are useful to ignore padding tokens. Mask value will be set to 0 for padding tokens and 1 for actual tokens. We will create mask both for training and test data

````python
# Create attention masks for training
train_attn_masks = []

# Create a mask of 1s for each token followed by 0s for padding
for seq in trainX:
  seq_mask = [float(i>0) for i in seq]
  train_attn_masks.append(seq_mask)
````

````python
# Create attention masks for Test
test_attn_masks = []

# Create a mask of 1s for each token followed by 0s for padding
for seq in testX:
  seq_mask = [float(i>0) for i in seq]
  test_attn_masks.append(seq_mask)
````

````python
print(train_attn_masks[100])
````

At this point, we have the data ready

## <font color='blue'>  Build the Model

````python
#Load Pre-trained Bert Model with a Binary Classification layer at the top.
#Huggingface library provides TFBertForSequenceClassification for the same
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
````

````python
# Checking the model summary and the layers and the number of parameters
model.summary()
````

````python
# Prepare training: Compile tf.keras model with optimizer, loss and learning rate schedule
optimizer = tf.keras.optimizers.Adam(learning_rate=3e-5, epsilon=1e-08, clipnorm=1.0)
````

````python
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
````

````python
metric = tf.keras.metrics.SparseCategoricalAccuracy('accuracy')
model.compile(optimizer=optimizer, loss=loss, metrics=[metric])
````

## <font color='blue'>  Train the Model

````python
import numpy as np
````

````python
train_x_data = {'input_ids': np.array(trainX), 'attention_mask': np.array(train_attn_masks)}
test_x_data = {'input_ids': np.array(testX), 'attention_mask': np.array(test_attn_masks)}
````

````python
model.fit(train_x_data, trainY,
          validation_data=(test_x_data, testY),
          batch_size=16,
          epochs=2)
````

The model is showing good results on the training and testing datasets, scoring 93% and 89% respectively, as can be seen from the results above. Also, the training and validation loss is relatively small as 0.17 and 0.27.
